# Countries with the Highest  and Lowest inflation Rate in 1961

```python
HInflation_df = inflation_df[['Country Name','Country Code','1961']].sort_values('1961', ascending= False).head()
HInflation_df
```

```python
LInflation_df = inflation_df[['Country Name','Country Code','1961']].sort_values('1961').head()
LInflation_df
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'heavy',
        'size': 16
        }
font2= {'family': 'serif',
        'color': 'blue',
        'weight': 'normal',
        'size': 14
        }

f, axs = plt.subplots(1, 2, figsize=(9,5),sharey = True)
plt.title('The Countries with The Highest and Lowest Inflation rate ', fontdict = font)

sns.barplot(x=HInflation_df['Country Code'], y = HInflation_df['1961'], ax= axs[0])
sns.barplot(x=LInflation_df['Country Code'], y = LInflation_df['1961'], ax= axs[1])
```

## Merging with a file that contains the Regions of the countries and each countries Income group and a special notes..

```python
#Read the csv file 
newfile=open('C:\\Users\\HP\\Desktop\\Metadata_Country_API_FP.CPI.TOTL.ZG_DS2_en_csv_v2_4250827.csv',"r")
Country_df= pd.read_csv(newfile)
Country_df
```

```python
#using the .merge function we merge the two dataframe
inflations_df = inflation_df.merge(Country_df, on='Country Code')
inflations_df
```

```python
#we renamed the columns, so as to avoid confusion....

inflations_df. rename(columns = {'Country Name':'Country_Name', 'Country Code':'Country_Code'}, inplace = True)
inflations_df
```

```python
inflations_df['IncomeGroup'].value_counts()
```

According to the dataset, We found out that the countries with High income - 29, countries with lower middle income - 13, also the countries with Upper middle income - 13, and the ones with lower income - 2 . 

## Countries with Upper Middle Income (UMI) 

```python
#We create a dataframe for the UMI Countries.
UMI = inflations_df[inflations_df['IncomeGroup'] == 'Upper middle income']
UMI
```

```python
#the Regions of the UMI countries in the datasets
UMI['Region'].value_counts()
```

```python
plt.figure(figsize=(16,8))
plt.title('The Upper middle Income Countries and their Regions', fontdict=font)

sns.barplot(x=UMI['2021'], y=UMI['Country_Name'], data=UMI, hue='Region')
plt.xlabel('Inflation Rate in 2021', fontdict=font2)
plt.ylabel('UMI Countries', fontdict=font2)
```

 ## Countries with High Income
